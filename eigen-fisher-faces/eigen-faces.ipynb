```python
import os
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
from random import randint

%matplotlib inline
```

```python
# Load face database (http://vision.ucsd.edu/content/yale-face-database).
#
# wget http://vision.ucsd.edu/datasets/yale_face_dataset_original/yalefaces.zip
# unzip yalefaces.zip
# mv yalefaces/Readme.txt .

face_db = 'yalefaces'
w = 320
h = 243

dir_path = os.path.join(os.getcwd(), face_db)
listed_images = os.listdir(dir_path)
num_faces = len(listed_images)
X = np.zeros((num_faces, w*h))
y = []

for idx, filename in enumerate(listed_images):
    X[idx, :] = data.load(os.path.join(dir_path, filename)).flatten()
    y.append(filename[0:filename.find('.')])
```

```python
# Compute PCA.
#
# Example:
# (1) Compute covariance. 
# (2) Compute eigenvectors and eigenvalues.
# (3) Sort eigenvectors with respect to eigenvalues in descending order.
# (4) Select first k columns of eigenvectors.
#
# There are two main cases that depends on size of input matrix X (= nxd).
# [1] n > d
# X  = 100x50
# n  = 100
# d  = 50
# mu = 1x50
# X_norm = X-mu (= 100x50)
# cov = X_norm'*X_norm (50x100 * 100x50 = 50x50)
#
# Perform step (2) and (3) and obtain V_sorted (= 50x50).
# Return the first k (up to 50) columns of V_sorted => 50xk.
#
# [2] n <= d
# X  = 100x10000
# n  = 100
# d  = 10000
# mu = 1x10000
# X_norm = X-mu (= 100x10000)
# cov = X_norm*X_norm' (100x10000 * 10000x100 = 100x100)
# 
# Perform step (2). You can see that computed eigenvectors does not explain all features (10000),
# but contain only 100 principal components (rows of input matrix X). With this we couldn't
# reconstruct original image using any number of principal components. 
#
# Therefore, we apply eigenvectors on normalized matrix X (X_norm) and obtain eigenvectors with
# required dimensions.
# V = X_norm'*V (10000x100 * 100x100 = 10000x100)
#
# This transformation keeps orthogonal property of eigenvector matrix but we require 
# eigenvector matrix to be orthonormal. Hence, each column of matrix has to be normalized.
# In the next step we perform (3) and then we can return the first k (up to 100) columns
# of V_sorted => 10000xk.
#
def pca(X, k):
    n, d = X.shape

    mu = np.mean(X, axis=0)
    X_norm = X - mu

    def sort_eigenvectors(w, V):
        sort_index = np.argsort(-w)
        V_sorted = V[:, sort_index]
        return V_sorted[:, 0:k]
    
    if n > d: # more faces than features (pixels)
        print "n > d"
        cov = np.dot(X_norm.T, X_norm)
        w, V = np.linalg.eig(cov)                      
    else:
        print "n <= d"
        cov = np.dot(X_norm, X_norm.T)        
        w, V = np.linalg.eig(cov)
        V = np.dot(X_norm.T, V)

        for i in range(0, n):
            V[:, i] = V[:, i]/np.linalg.norm(V[:, i])                
        
    return sort_eigenvectors(w, V), mu
```

```python
# Normalize vector to values between low and high boundaries.
def normalize(X, low, high):
    X_min = np.min(X)
    X_max = np.max(X)
    
    X_norm = (X - X_min)/(X_max - X_min)    
    X_norm = X_norm * (high-low)
    X_norm = X_norm + low
    
    return X_norm

def to_grayscale(X, height, width):
    X_gray = normalize(X.astype(np.float16), 0.0, 255.0)
    return X_gray.reshape((height, width)).astype(np.uint8)
```

```python
# Perform PCA.
k = 30 # number of principal components
V_old, mu_old = pca(X[0:100, :], k)

print V_old.shape
```

```python
# Display Eigenfaces.
def display_eigen_face(V, h, w, rows, cols, pos):
    fig.add_subplot(rows, cols, pos+1)
    plt.imshow(to_grayscale(V[:, pos], h, w), cmap='jet')
    plt.title("Principal component number " + str(pos+1))
    plt.colorbar()
    
fig = plt.figure(figsize=(20, 20))
for i in range (0, 4):
    display_eigen_face(V_old, h, w, 2, 2, i)
```

```python
# Detect faces.
import cv2
from skimage import transform

faceCascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

h_face = 100
w_face = 100

X_faces = np.zeros((num_faces, h_face*w_face))

for i in range(0, X.shape[0]):
    img = X[i, :].reshape((h, w)).astype(np.uint8)
    
    # expect only one face inside an image
    r = faceCascade.detectMultiScale(img, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30), flags = cv2.CASCADE_SCALE_IMAGE)[0]
    
    X_faces[i, :] = transform.resize(img[r[1]:r[1]+r[3], r[0]:r[0]+r[2]], (h_face, w_face)).reshape((1, h_face*w_face))
```

```python
# Display random faces.    
fig = plt.figure(figsize=(10, 10))
for i in range(1, 10):
    display_random_face(X_faces, 3, 3, i, h_face, w_face)
```

```python
# Perform PCA again
k = 300 # number of principal components
V_new, mu_new = pca(X_faces, k)

print V_new.shape
```

```python
# Plot new principal components.
fig = plt.figure(figsize=(20, 20))
for i in range(0, 9):
    display_eigen_face(V_new, h_face, w_face, 3, 3, i)
```

```python
# Project from original space to subdimensional one.
# y = (x - mu)*V
#
# Example (10 principal components, 10000 pixels in image):
# V  = 10000x10
# X  = 1x10000
# mu = 1x10000
# 
# First normalize and then project to subdimensional space.
# X*V = 1x10
#
# Image was downsampled from 1x10000 to 1x10, that is 1000 times 
# smaller than image in original space.
def project(V, X, mu):
    return np.dot(X - mu, V)

# Reconstruct from subdimensional space to origional one.
# X = y*V' + mu
#
# Example (10 principal components, 10000 pixels in image):
# V  = 10000x10
# y  = 1x10
# mu = 1x10000
#
# First, project to original space.
# y*V' = 1x10000
#
# Bias features in original space with precomputed mean. Image is now
# represented with the same number of features as before projection
# to subdimensional space.
def reconstruct(V, y, mu):
    return np.dot(y, V.T) + mu

# Reconstruct random image
fig = plt.figure(figsize=(20, 20))
rand_idx = randint(0, num_faces)
print rand_idx
rand_img = X_faces[randint(0, num_faces), :]
cmap = 'gray' # 'jet'

for eigen_idx in range(1, 26):
    P = project(V_new[:, 0:eigen_idx], rand_img, mu_new)
    R = reconstruct(V_new[:, 0:eigen_idx], P, mu_new)
    tmp_img = to_grayscale(R, h_face, w_face)
    
    fig.add_subplot(5, 5, eigen_idx)
    plt.imshow(tmp_img, cmap=cmap)
    plt.title(str(eigen_idx) + " Eigenvectors")    
```

```python
# Face recognition
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing
from sklearn.model_selection import LeaveOneOut

loo = LeaveOneOut()
scores = []

n_neighbors=3

# Preprocess labels.
le = preprocessing.LabelEncoder()
y_labels = le.fit_transform(y)
pca_idx_max = 50

for pca_idx in range(1, pca_idx_max):
    # Project all faces to subdimensional feature space.
    P_faces = project(V_new[:, 0:pca_idx], X_faces, mu_new)

    scores_loo = []

    for train, test in loo.split(range(0, num_faces)):        
        # Predict with k-NN
        neigh = KNeighborsClassifier(n_neighbors=n_neighbors)
        neigh.fit(P_faces[train, :], y_labels[train])

        # Compute score on training dataset
        scores_loo.append(neigh.score(P_faces[test, :], y_labels[test]))
        
    scores.append(np.mean(scores_loo))
    
print max(scores)
plt.figure(figsize=(5, 5))
plt.plot(range(1, pca_idx_max), scores)
plt.xlabel("# principal components")
plt.ylabel("mean accuracy")
plt.show()
```
