```python
import os
from random import randint
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
from sklearn import preprocessing

%matplotlib inline
```

```python
# Load face database (http://vision.ucsd.edu/content/yale-face-database)
# wget http://vision.ucsd.edu/datasets/yale_face_dataset_original/yalefaces.zip
# unzip yalefaces.zip
# mv yalefaces/Readme.txt .

face_db = 'yalefaces'
w = 320
h = 243

dir_path = os.path.join(os.getcwd(), face_db)
listed_images = os.listdir(dir_path)
num_faces = len(listed_images)
X = np.zeros((num_faces, w*h))
y = []

for idx, filename in enumerate(listed_images):
    X[idx, :] = data.load(os.path.join(dir_path, filename)).flatten()
    y.append(filename[0:filename.find('.')])
    
# Detect faces.
import cv2
from skimage import transform

faceCascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

h_face = 100
w_face = 100

X_faces = np.zeros((num_faces, h_face*w_face))

for i in range(0, X.shape[0]):
    img = X[i, :].reshape((h, w)).astype(np.uint8)
    
    # expect only one face inside an image
    r = faceCascade.detectMultiScale(img, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30), flags = cv2.CASCADE_SCALE_IMAGE)[0]
    
    X_faces[i, :] = transform.resize(img[r[1]:r[1]+r[3], r[0]:r[0]+r[2]], (h_face, w_face)).reshape((1, h_face*w_face))
```

```python
# Taken from Eigenfaces example. Nothing was changed.
#
# Compute PCA.
def pca(X, k):
    n, d = X.shape

    mu = np.mean(X, axis=0)
    X_norm = X - mu

    def sort_eigenvectors(w, V):
        sort_index = np.argsort(-w)
        V_sorted = V[:, sort_index]
        return V_sorted[:, 0:k]
    
    if n > d: # more faces than features (pixels)
        cov = np.dot(X_norm.T, X_norm)
        w, V = np.linalg.eig(cov)                      
    else:
        cov = np.dot(X_norm, X_norm.T)        
        w, V = np.linalg.eig(cov)
        V = np.dot(X_norm.T, V)

        for i in range(0, n):
            V[:, i] = V[:, i]/np.linalg.norm(V[:, i])                
        
    return sort_eigenvectors(w, V), mu

# Project from original space to subdimensional one.
def project(V, X, mu):
    return np.dot(X - mu, V)

# Reconstruct from subdimensional space to origional one.
def reconstruct(V, y, mu):
    return np.dot(y, V.T) + mu

# Normalize vector to values between low and high boundaries.
def normalize(X, low, high):
    X_min = np.min(X)
    X_max = np.max(X)
    
    X_norm = (X - X_min)/(X_max - X_min)    
    X_norm = X_norm * (high-low)
    X_norm = X_norm + low
    
    return X_norm

def to_grayscale(X, height, width):
    X_gray = normalize(X.astype(np.float16), 0.0, 255.0)
    return X_gray.reshape((height, width)).astype(np.uint8)
```

```python
# Compute LDA (Linear Discriminant Analysis).
#
# Example:
# X = 166x10000
# y = 166x1
# k = 50
#
# d   = 10000
# S_b = 10000x10000 (too big!)
# S_w = 10000x10000 (too big!)
#
# Iterate through all classes and compute between-class (S_b) and within-class (S_w)
# scatter matrices whose ratio we want to maximize => argmax( W'*S_b*W / W'*S_w*W ).
#
# Scatter matrices are composed by a sum of class dependant scatter matrices. Between-class
# scatter matrix is represented by a sum of n (number of samples in class) square differences
# between class-dependant and global means. Within-class matrix is represented by a sum of 
# covariance matrices computed from normalized features within class. Maximization problem 
# can be solved by general eigenvalue problem: 
#   S_b*V_i = lambda_i*S_w*V_i
#   S_w^(-1)*S_b*V_i = lambda_i*V_i
#
# Obtained eigenvectors are sorted (in descending manner) according to related eigenvalues.
# Function returns the first k components (= first k sorted eigenvectors) and global mean.
#
# V_sorted = 10000xk(=50)
#
def lda(X, y, k):
    _, d = X.shape
    
    # labels preprocessing
    labels = np.unique(y)
    C = len(labels)
    
    # allocate S_w and S_b
    S_w = np.zeros((d, d))
    S_b = np.zeros((d, d))
    
    mu = np.mean(X, axis=0)
    
    for i in range(0, C):
        X_i = X[y == i, :]        
        n, _ = X_i.shape
        
        # normalize data
        mu_i = np.mean(X_i, axis=0)
        X_i_norm = X_i - mu_i
        
        # within-class scatter
        S_w = S_w + np.dot(X_i_norm.T, X_i_norm)
        
        # between-class scatter
        S_b = S_b + n * np.dot((mu_i-mu).T, (mu_i-mu))
        
    w, V = np.linalg.eig(np.dot(np.linalg.inv(S_w), S_b))
    
    def sort_eigenvectors(w, V, k):
        sort_index = np.argsort(-w)
        V_sorted = V[:, sort_index]
        return V_sorted[:, 0:k]
        
    return sort_eigenvectors(w, V, k), mu
```

```python
# Fisherfaces use internally both LDA and PCA in order to downsample 
# dimension of given features.
#
# Example:
# X = 166x10000
# y = 166x1
# k = 30
#
# n = 166
# C = 15
# k = 15-1 = 14
# k = min(k, C-1) = min(30, 14) = 14
# 
# W_pca = 10000x(166-15 = 151)
# PCA projection of X => 166x151
# W_lda = 151x14
# W = W_pca * W_lda = 10000x151 * 151x14 = 10000x14
#
def fisherfaces(X, y, k):
    n, _ = X.shape
    labels = np.unique(y)
    C = len(labels)    
    
    # expect that number of records in dataset is larger than number of classes
    W_pca, mu = pca(X, n-C) # rank of matrix is at most n-C
    W_lda, mu_lda = lda(project(W_pca, X, mu), y, min(k, C-1)) # keep at most C-1 eigenvectors
    W = np.dot(W_pca, W_lda)
    
    return W, mu
```

```python
# Compute Fisherfaces.
k = 30 # is larger than C-1 => 30 principal components shall not be used
le = preprocessing.LabelEncoder()
y_labels = le.fit_transform(y)
V, mu = fisherfaces(X_faces, y_labels, k)
```

```python
# Display Fisherfaces.
def display_fisher_face(V, h, w, rows, cols, pos):
    fig.add_subplot(rows, cols, pos+1)
    plt.imshow(to_grayscale(V[:, pos], h, w), cmap='jet')
    plt.title("Fisherface " + str(pos+1))
    plt.colorbar()
    
fig = plt.figure(figsize=(20,20))
for i in range (0, V.shape[1]):
    display_fisher_face(V, h_face, w_face, 4, 4, i)
```

```python
# Reconstruct random image.
fig = plt.figure(figsize=(20, 20))
rand_idx = randint(0, num_faces)
print rand_idx
rand_img = X_faces[rand_idx, :]

cmap = 'gray' # 'jet'

for fisherface_idx in range(0, V.shape[1]):
    P = project(V[:, 0:fisherface_idx], rand_img, mu)
    R = reconstruct(V[:, 0:fisherface_idx], P, mu)
    tmp_img = to_grayscale(R, h_face, w_face)
    
    fig.add_subplot(5, 5, fisherface_idx+1)
    plt.imshow(tmp_img, cmap=cmap)
    plt.title(str(fisherface_idx+1) + " Fisherfaces")    
```

```python
# Face recognition
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import LeaveOneOut

scores = []
loo = LeaveOneOut()
n_neighbors = 5

fisher_idx_max = V.shape[1]

for fisher_idx in range(1, fisher_idx_max):
    # Project all faces to subdimensional feature space.
    P_faces = project(V[:, 0:fisher_idx], X_faces, mu)    
    
    scores_loo = []

    for train, test in loo.split(range(0, num_faces)):
        # Predict with k-NN
        neigh = KNeighborsClassifier(n_neighbors=n_neighbors)
        neigh.fit(P_faces[train, :], y_labels[train])

        # Compute score on training dataset
        scores_loo.append(neigh.score(P_faces[test, :], y_labels[test]))
    
    scores.append(np.mean(scores_loo))
    
print max(scores)
plt.figure(figsize=(5, 5))
plt.plot(range(1, fisher_idx_max), scores)
plt.xlabel("# Fisherfaces")
plt.ylabel("mean accuracy")
plt.show()
```
